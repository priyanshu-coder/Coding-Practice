Given an integer matrix of size m*n, you need to find out the value of minimum cost to reach from the cell (0, 0) to (m-1, n-1).
From a cell (i, j), you can move in three directions : (i+1, j), (i, j+1) and (i+1, j+1).
Cost of a path is defined as the sum of values of each cell through which path passes.

```python
#recursive
import sys
def findCost(li,i,j,m,n):
    
    if i==m-1 and j==n-1:
        return li[i][j]
    
    if i>=m or j>=n:
        return sys.maxsize
    
    cost=li[i][j]+min(findCost(li,i+1,j,m,n),findCost(li,i,j+1,m,n),findCost(li,i+1,j+1,m,n))
    return cost
    
    
li=[int(x) for x in input().split()]

m=li[0]
n=li[1]
newLi=[]
for _ in range(m):
    ll=[int(y) for y in input().split()]
    newLi.append(ll)
    
mm=len(newLi)
nn=len(newLi[0])
print(findCost(newLi,0,0,mm,nn))
```

```python
#memoization
#memoization
import sys
def findCost(li,i,j,m,n,dp):
    
    if i==n-1 and j==m-1:
        return li[i][j]
    
    if i>=n or j>=m:
        return sys.maxsize
    
    if dp[i][j+1]==sys.maxsize:
        ans1=findCost(li,i,j+1,m,n,dp)
        dp[i][j+1]=ans1
        
    else:
        ans1=dp[i][j+1]
        
    if dp[i+1][j]==sys.maxsize:
        ans2=findCost(li,i+1,j,m,n,dp)
        dp[i+1][j]=ans2
        
    else:
        ans2=dp[i+1][j]
        
    if dp[i+1][j+1]==sys.maxsize:
        ans3=findCost(li,i+1,j+1,m,n,dp)
        dp[i+1][j+1]=ans3
        
    else:
        ans3=dp[i+1][j+1]
        
    cost=li[i][j]+min(ans1,ans2,ans3)
        
    return cost


li=[int(x) for x in input().split()]

m=li[0]
n=li[1]
newLi=[]
for _ in range(m):
    ll=[int(y) for y in input().split()]
    newLi.append(ll)
    
mm=len(newLi)
nn=len(newLi[0])
dp=[[sys.maxsize for _ in range(nn+1)] for _ in range(mm+1)]
print(findCost(newLi,0,0,mm,nn,dp))
```

```python
#iteration
import sys
def findCost(li,m,n):
    
    dp=[[sys.maxsize for _ in range(n+1)] for _ in range(m+1)]
    
    
    for i in range(m-1,-1,-1):
        for j in range(n-1,-1,-1):
            if i==m-1 and j==n-1:
                dp[i][j]=li[m-1][n-1]
                
            else:
                
                ans1=dp[i+1][j]
                ans2=dp[i][j+1]
                ans3=dp[i+1][j+1]
                dp[i][j]=li[i][j]+min(ans1,ans2,ans3)
                
    return dp[0][0]
    
li=[int(x) for x in input().split()]

m=li[0]
n=li[1]
newLi=[]
for _ in range(m):
    ll=[int(y) for y in input().split()]
    newLi.append(ll)
    
mm=len(newLi)
nn=len(newLi[0])
print(findCost(newLi,mm,nn))
```
