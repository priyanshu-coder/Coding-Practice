```python
import matplotlib.pyplot as plt
```

```python
#works as y values
plt.plot([2,4,6,8])
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("my graph")
plt.grid()
plt.show()
```

```python
#x axis values then y axis values
plt.plot([2,4,6,8],[1,2,3,4],"ro")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("my graph")
plt.grid()
plt.show()
```

```python
import numpy as np
```

```python
t=np.arange(0.,5.,0.2)
```

```python
plt.plot(t,t**2, "r--", label="^2", linewidth=5.0)
plt.plot(t,t**2.2, "bs", label="^2.2")
plt.plot(t,t**2.5, "g^", label="^2.5")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("my graph")
plt.legend()#adds labels
plt.show()
```

```python
x1=[1,2,3,4]
x2=[1.5,2.5,3.5,4.5]
y1=[2,4,6,8]
y2=[2.5,3.6,4.66,7.55]

line=plt.plot(x1,y1,x2,y2)
plt.setp(line[0],color='r',linewidth=2.0)
plt.setp(line[1],color='b' ,linewidth=2.0)
plt.show()
```

```python
t1 = np.arange(0.,5.,0.2)
t2 = np.arange(0.,5.,0.02)

def something(inp):
    return np.cos(2*np.pi*inp)*np.exp(-inp)

plt.figure(1)
plt.subplot(211)
plt.plot(t1,something(t1))
plt.subplot(212)
plt.plot(t2,something(t2),'r--')
plt.grid()
plt.show()
```
